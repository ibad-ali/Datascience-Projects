```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\Ibad\Desktop\DT.csv")
df
```

```python
new_df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked', 'Survived'], axis = 'columns')
new_df
```

```python
target = df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_sex = LabelEncoder()
```

```python
new_df['Sex_n'] = lab_sex.fit_transform(new_df['Sex']) 
#since DT algo is working on only numbers for
#this convert our SEX column values in to integer 1 for male and 0 for female
new_df
```

```python
new_df = new_df.drop(['Sex'], axis='columns') #now drop old column of Sex
new_df
```

```python
new_df['Age'].fillna(new_df['Age'].mean(), inplace = True)
```

```python
new_df
```

```python
 from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(new_df, target)
```

```python
model.predict([[1,32,76.2917,0]]) #random row predict
```

```python
model.score(new_df, target) #here is accuracy 100% beacuse i train and test on same data
```
