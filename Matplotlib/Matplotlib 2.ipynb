```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
country = ["USA", "UAE", "NZ", "PAK", "CHINA"]
population = [1.5,1.6, 1.75, 1.8, 2.5]
size = [1, 1.4, 1.3, 1.6, 2]
```

```python
plt.bar(country, population, label="Population") #for vertical
plt.legend(loc="best")
```

```python
plt.barh(country, population, label="Population") #for horizontal
plt.legend(loc="best")
```

```python
blood = [113,100,45,65,58,75,90,87,113,95,145,125,91,130,135,111,99,141]
plt.hist(blood, bins=[45,75,105,145], rwidth=0.95, color="yellow") #bins means range 
```

```python
#two list histogram side by side
blood = [113,100,45,65,58,75,90,87,113,95,145,125,91,130,135,111,99,141]
sugar = [110,90,55,45,75,53,95,81,103,88,85,145,115,130,105,101,90,61]

plt.hist([blood,sugar], bins=[45,75,105,145], rwidth=0.95, color=["purple","orange"], label=["blood","sugar"]) #for making horizontal write orientation = "horizontal"
plt.legend(loc="upper left")
```

```python
amount = [1400, 600, 400, 250, 100, 200]
exp_labels = ["Home Rent", "Food", "Mobile", "Bills", "Fees", "Others"]
plt.pie(amount,labels=exp_labels, radius = 1.5, autopct="%0.1f%%", explode=[0,0,0.2,0,0,0.2], startangle=45)
#radius control the size of pie chart
#auto pct id for percentage of every block in piechart 0.1 define after decimal one value return
#explode is for break any block of piechart the list length is always equal to the total no of blocks in piechart
#start angle is for change the rotation of piechart
plt.savefig("piechart.png", bbox_inches = "tight") #for saving the pictures
```
