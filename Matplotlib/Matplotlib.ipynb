```python
#line Plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = np.arange(1,11)
y = 2 * x
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.show()
```

```python
plt.plot(x,y, color='g', linestyle=':', linewidth=3)
plt.show()
```

```python
# two lines in graph
y1 = x*2
y2 = x*3
plt.plot(x,y1,color='g')
plt.plot(x,y2,color='r')
plt.grid(True)
plt.show()
```

```python
#Sub Plots
plt.subplot(1,2,1) #Total row=1, Total column=2, cell=1
plt.plot(x,y1,color='g')
plt.subplot(1,2,2) #Total row=1, Total column=2, cell=2
plt.plot(x,y2,color='r')
plt.show()
```

```python
#Sub Plots
plt.subplot(2,1,1) #Total row=1, Total column=2, cell=1
plt.plot(x,y1,color='g')
plt.subplot(2,1,2) #Total row=1, Total column=2, cell=2
plt.plot(x,y2,color='r')
plt.show()
```

```python
#Bar Plot
students = {'Ibad':87, 'Hamza':62, 'Ali':75, 'Usman':55}
names = list(students.keys())
marks = list(students.values())
plt.bar(names,marks)
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
plt.barh(names,marks,color='r')
plt.xlabel('Marks')
plt.ylabel('Names')
plt.show()
```

```python
#Scatter Plot
x = [10, 20, 30, 40, 50, 60]
a = [51, 62, 44, 72, 10, 59]
plt.scatter(x,a)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60]
a = [51, 62, 44, 72, 10, 59]
plt.scatter(x,a,c='r', marker='*',s=100)
plt.show()
```

```python
#Histogram
data = [6,8,1,6,3,1,2,4,1,5,9,8,2,3,5]
plt.hist(data)
plt.show() 
```

```python
data = [6,8,1,6,3,1,2,4,1,5,9,8,2,3,5]
plt.hist(data, color='g', bins=5)
plt.show()
```

```python
# Pie Chart
names = ['Physics', 'Chemistry', 'Maths', 'Computer']
area = [100, 55, 85, 62]
plt.pie(area, labels=names)
plt.show()
```
