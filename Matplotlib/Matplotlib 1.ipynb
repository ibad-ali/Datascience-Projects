```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [23,22,35,25,28,32,19]
```

```python
plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Weather Graph")
plt.plot(x,y, linestyle="dotted", color="black", linewidth=3) 
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html?highlight=plot#matplotlib.pyplot.plot
```

```python
plt.plot(x,y, "g*--") #(g=green, *=points, --=line style) we use it any arrangement ie *g--, --*g
#for more design and formats 
#link:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html?highlight=plot#matplotlib.pyplot.plot
```

```python
plt.plot(x,y,color="green", linewidth=2, alpha=0.5) #alpha is for transparency and it value is between 0 - 1
```

```python
days = [1,2,3,4,5,6,7]
max_temp = [50,51,52,48,47,49,46]
min_temp = [43,42,40,44,33,35,37]
avg_temp = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_temp, label="Max")
plt.plot(days,min_temp, label="Min")
plt.plot(days,avg_temp, label="Average")
plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Weather Graph")
plt.legend(loc="best", shadow=True) #loc means location of legend best means it put it where it found blank space
plt.grid() #to analyze the datapoints easily
```
