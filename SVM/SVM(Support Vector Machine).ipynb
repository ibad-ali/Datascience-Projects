```python
import pandas as pd
from sklearn.datasets import load_digits
digit = load_digits()
```

```python
dir(digit)
```

```python
df = pd.DataFrame(digit.data)
df
```

```python
df.columns
```

```python
df['target'] = digit.target
```

```python
df.head(15)
```

```python
digit.target_names #no of values in target variables
```

```python
df[df.target == 9] 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.2)
# here i import train test module and by using this module train our model by 80% row of all features before (,)
#train columns and after target 
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

```python
len(X_train) #80 percent rows
```

```python
len(X_test) #20 percent rows
```

```python
model.score(X_test, Y_test)
```
