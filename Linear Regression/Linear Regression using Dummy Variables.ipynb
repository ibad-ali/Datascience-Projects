```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:\\Users\Ibad\Desktop\car.csv")
df
```

```python
dummies = pd.get_dummies(df.Model)
dummies
```

```python
con = pd.concat([df,dummies], axis = 'columns')
con
```

```python
final = con.drop(['Model'], axis = 'columns')
final
```

```python
ful_final = final.drop(['Price'], axis= 'columns')
x = ful_final
x
```

```python
y = final.Price
y
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[86000,7,0,1,0]])
```

```python
model.predict([[45000,4,0,0,1]])
```

```python
model.score(x,y) #check model accuracy 
```
